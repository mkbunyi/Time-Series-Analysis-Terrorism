```python
from pandas import read_csv
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN, Bidirectional, GRU, LSTM
import keras.regularizers
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import math
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Parameter split_percent defines the ratio of training examples
def get_train_val_test(url, colno=1,train_percent=0.8,val_percent=0.1):
    df = read_csv(url, usecols=[colno], engine='python')
    print(df)
    
    data = np.array(df.values.astype('float32'))
    n = len(data)
    
    #split points
    training_index = int(n*train_percent)
    val_index = int(n*(train_percent+val_percent))

    #Normalize the data using training set
    mean = data[:training_index].mean(axis=0)
    data -= mean
    std = data[:training_index].std(axis=0)
    data /= std

    # split into train, validation, and test sets
    train_data = data[range(training_index)]
    val_data = data[training_index:val_index]
    test_data = data[val_index:]
    
    # Split 
    return train_data, val_data, test_data, data
```

```python
# Prepare the input X and target Y
def get_XY(dat, time_steps):
    # Indices of target array
    Y_ind = np.arange(time_steps, len(dat), time_steps)
    Y = dat[Y_ind]
    # Prepare X
    rows_x = len(Y)
    X = dat[range(time_steps*rows_x)]
    X = np.reshape(X, (rows_x, time_steps, 1))    
    return X, Y
```

```python
def create_model(modtype,hidden_units, dense_units, input_shape, activation):
    # input_shape specifies the parameter (time_steps x features)
    model = Sequential()
    if modtype == 'RNN':    
        model.add(SimpleRNN(hidden_units, input_shape=input_shape, 
                            activation=activation[0]))
    if modtype == 'GRU':
        model.add(GRU(hidden_units, input_shape=input_shape, 
                            activation=activation[0]))
    if modtype == 'LSTM':
        model.add(LSTM(hidden_units, input_shape=input_shape, 
                            activation=activation[0]))
    model.add(Dense(units=dense_units, activation=activation[1]))
    model.compile(loss='mean_squared_error', optimizer='adam')
    #model.compile(optimizer="RMSprop", loss='mae') # MAE makes this comaprable with common sense approach
    return model
```

```python
def create_model_reg(modtype,hidden_units, dense_units, input_shape, activation):
    # input_shape specifies the parameter (time_steps x features)
    model = Sequential()
    if modtype == 'RNN':    
        model.add(SimpleRNN(hidden_units, input_shape=input_shape, 
                            activation=activation[0], kernel_regularizer=keras.regularizers.l2(l=0.1)))
    if modtype == 'GRU':
        model.add(GRU(hidden_units, input_shape=input_shape, 
                            activation=activation[0], kernel_regularizer=keras.regularizers.l2(l=0.1)))
    if modtype == 'LSTM':
        model.add(LSTM(hidden_units, input_shape=input_shape, 
                            activation=activation[0], kernel_regularizer=keras.regularizers.l2(l=0.1)))
    model.add(Dense(units=dense_units, activation=activation[1]))
    model.compile(loss='mean_squared_error', optimizer='adam')
    #model.compile(optimizer="RMSprop", loss='mae') # MAE makes this comaprable with common sense approach
    return model
```

```python
def print_error(trainY, testY, train_predict, test_predict):    
    # Error of predictions
    train_rmse = math.sqrt(mean_squared_error(trainY, train_predict))
    test_rmse = math.sqrt(mean_squared_error(testY, test_predict))
    # Print RMSE
    print('Train RMSE: %.3f RMSE' % (train_rmse))
    print('Test RMSE: %.3f RMSE' % (test_rmse))    
```

```python
# Plot the result
def plot_result(trainY, testY, train_predict, test_predict):
    actual = np.append(trainY, testY)
    predictions = np.append(train_predict, test_predict)
    rows = len(actual)
    plt.figure(figsize=(15, 6), dpi=80)
    plt.plot(range(rows), actual)
    plt.plot(range(rows), predictions)
    plt.axvline(x=len(trainY), color='r')
    plt.legend(['Actual', 'Predictions'])
    plt.xlabel('Observation number after given time steps')
    plt.ylabel('Normalized Monthly Terrorism Incident Count')
    plt.title('Monthly Terrorism Incident Count: Actual and Predicted Values')
```

```python
train_data, val_data, test_data, data = get_train_val_test('gtd_monthly.csv',colno=2,val_percent=0)

time_steps = 3 #1 #3 #6 #12 #24 #36
trainX, trainY = get_XY(train_data, time_steps)
valX, valY = get_XY(val_data, time_steps)
testX, testY = get_XY(test_data, time_steps)
```

```python
#def create_model(modtype,hidden_units, dense_units, input_shape, activation):
model = create_model('RNN',2, 1, (time_steps,1), activation=['linear', 'linear'])
# make predictions
train_predict = model.predict(trainX)
test_predict = model.predict(testX)
# Mean square error
print_error(trainY, testY, train_predict, test_predict)
```

```python
plot_result(trainY, testY, train_predict, test_predict)
```

```python
#def create_model(modtype,hidden_units, dense_units, input_shape, activation):
model = create_model_reg('RNN',2, 1, (time_steps,1), activation=['linear', 'linear'])
# make predictions
train_predict = model.predict(trainX)
test_predict = model.predict(testX)
# Mean square error
print_error(trainY, testY, train_predict, test_predict)
```
